```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data.info()
```

```python
data.head(92)
```

```python
data["input[V]"] = [i[1::] for i in data["input[V]"]]
data["output[V]"] = [i[1::] for i in data["output[V]"]]
```

```python
data = data [::-1]
```

```python
data = data [1::]
data.head()
```

```python
data.tail()
```

```python
widths = ["20", "30", "40", "50", "60", "70", "80"]
voltages = list(data["voltage_level[V]"].unique())
```

```python
drop_list = []
for volt in voltages:
    filter1 = data["voltage_level[V]"] == volt
    data1 = data[filter1]
    for width in widths:
        filter2 = data1["width[percent]"] == int(width)
        data2 = data1[filter2]
        index_list = list(data2.index.values)
        #print(index_list)
        for j in range(4):
            drop_list.append(index_list[j])
clean_data = data.drop(drop_list)
```

```python
clean_data.head(260)
```

```python
features = ["time", "voltage_level[V]", "width[percent]", "input[V]", "output[V]"]
```

```python
dictionary = {"time":clean_data["time"], "voltage_level[V]":clean_data["voltage_level[V]"], 
              "width[percent]":clean_data["width[percent]"], "input[V]":clean_data["input[V]"], 
              "output[V]":clean_data["output[V]"]}
```

```python
df = pd.DataFrame(dictionary, columns = features) 
```
